## Dataset: NFL Game Data

Import packages, load data, and concatenate yearly datasets:

```python
import numpy as np
import pandas as pd
import kagglehub

import seaborn as sns
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = (9, 5)

from scipy.stats import ttest_ind
```

```python
# Download latest version
path = "C:/Users/jadla/Classes/COGS108/data/"
#path = "/Users/frees/Desktop/COGS 108/data/"

nfl_games = []
for year in range(2010, 2020):
    file_path = f'{path}game_stats_{str(year)}.csv'
    nfl_year = pd.read_csv(file_path)
    nfl_year['season'] = year
    nfl_games.append(nfl_year)

nfl = pd.concat(nfl_games, ignore_index=True)
nfl
```

```python
def clean(df):

    out = df.copy()
    
    out['HomeWin'] = (out['H-Score'] > out['A-Score']).astype(int)
    out.columns = out.columns.str.replace('-', '_').str.lower()
    out['betting_over'] = out['result'].replace(-1, 0)
    out = out.drop(['result'], axis=1)
    out['h_totalyards'] = out['h_passyards'] + out['h_rushyards']
    out['a_totalyards'] = out['a_passyards'] + out['a_rushyards']
    out['hometeam'] = out['hometeam'].replace('SDG', 'LAC')
    out['awayteam'] = out['awayteam'].replace('SDG', 'LAC')
    out['hometeam'] = out['hometeam'].replace('STL', 'LAR')
    out['awayteam'] = out['awayteam'].replace('STL', 'LAR')
    out['score_diff'] = out['h_score'] - out['a_score']

    return out

nfl_clean = clean(nfl)
nfl_clean.head()
```

```python
# Check for null values
nfl_clean.isnull().sum()
```

As seen above, our cleaned dataset contains zero null values.

### Section I: Analysis of Home Team Score vs. Away Team Score

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = (9, 5)
```

```python
plt.figure()
sns.histplot(data=nfl_clean, x='h_score', color='darkcyan')
plt.xlabel("Home Score")
plt.title('Histogram of Home Team Score')

plt.figure()
sns.histplot(data=nfl_clean, x='a_score', color='crimson')
plt.xlabel("Away Score")
plt.title('Histogram of Away Team Score')

plt.show()
```

These two histograms display the distribution of points scored by home teams and away teams across all games in the dataset. The x-axis of each plot represents the number of points scored in a game, while the y-axis shows how many games ended with that particular score, separately for home and away teams. Both distributions are similarly shaped, with most scores clustering between roughly 10 and 35 points. However, home teams tend to score slightly higher than away teams: the home score histogram is shifted a bit to the right compared to the away score histogram. This suggests that home teams, on average, have a scoring advantage. High-scoring games (over 40 points) are infrequent for both groups, but slightly more common for home teams. The similarity in shape but difference in central tendency provides early evidence for a home-field advantage in scoring, which can be further investigated by examining score differentials and win percentages.

```python
plt.figure()
sns.histplot(x='score_diff', data=nfl_clean, color='darkslateblue')
plt.title('Histogram of Score Differential')
plt.xlabel('Score Differential (Home Score - Away Score)')
plt.show()
```

This histogram illustrates the distribution of score differentials (home score minus away score) for all games. The x-axis is the score differential, and the y-axis is the count of games with each differential. The distribution is centered slightly above zero, with a mean score differential of about +2.2 in favor of the home team. Most games have a differential between -20 and +20, but there are outliers on both sides.
```python
# hypothesis test for home/away score
h_score = nfl_clean['h_score']
a_score = nfl_clean['a_score']

test = ttest_ind(h_score, a_score)
t_stat = test.statistic
p_val = test.pvalue

t_stat, p_val
```

The t-test above compares the distribution of score for home teams and away teams. The test statistic of 7.73 and corresponding p-value of 1.27e-14 show that there is a statistically significant difference between the two distributions. The test statistic provides substantially strong evidence that home teams, on average, score more points per game than away teams.

The findings demonstrate a clear and consistent performance boost across various aspects of the sport for teams playing in their home stadiums.


The most direct evidence of this advantage is in the win outcomes. Home teams won approximately 56.5% of the games we studied and scored, on average, 2.2 more points than their opponents. Given the large number of games across a wide selection of years, this scoring advantage is likely not an anomaly and can be extrapolated to other time frames. This aligns well with our prior research on the subject and quantifies the advantage with a modern, data-rich timeframe. Our analysis gives a nuanced view by showing that this advantage extends deep into a team’s offensive performance. Home teams have been shown to consistently gain more total yards, passing yards, and rushing yards than their opponents. This could suggest that factors like crowd support and environment familiarity directly translate to a more effective offense.

In terms of yardage, home teams consistently gained more total yards than away teams in every season from 2010 to 2019. This finding reflects both slightly higher rushing and passing yards on average for home teams. The data visualizations reinforce this: passing yardage trends show home teams outperforming away teams across most seasons, especially in peak passing years like 2015 and 2016. Similarly, rushing yardage favors home teams, with a particularly sharp contrast during the mid-decade seasons. These results suggest that playing at home may lead to better offensive execution, possibly due to better communication, reduced travel strain, or weather familiarity.

When analyzing rushing attempts, we observe that home teams tended to rush more frequently than away teams in most seasons. This may suggest greater control of game flow and confidence in offensive planning when playing at home. It's plausible that teams feel more comfortable running the ball with crowd support and no travel fatigue, especially in later-game situations when running becomes more strategic.

It could be of interest that the one area where home-field advantage did not appear to have a clear impact was turnovers. Different from all other metrics, the data showed no consistent pattern favoring either the home or away team for turnovers. This might suggest that while playing at home may increase a team’s offensive confidence and rhythm, it may not be as effective in preventing mistakes (or forcing mistakes in the opponent). This complicates the popular narrative that crowd noise is the primary driver of home-field advantage. Perhaps players have adjusted to the difference in efficiency and subconsciously change their playstyles to be more or less risk-adverse if they are at home or away.

Taken together, our findings support the existence of a measurable home-field advantage in the NFL, particularly in scoring, total yards, and offensive rhythm, but not in all aspects of the game. The absence of a turnover effect, for example, shows that not every game metric is equally sensitive to the environment.

This research extends previous work by offering a more granular view into how home-field advantage operates. Rather than just confirming its existence, we’ve examined which parts of the game are most affected. Future studies could dig deeper into situational data—e.g., 4th-quarter performance, penalty differentials, or referee decisions—especially with more context such as crowd size or stadium design.

In conclusion, the home-field advantage in the NFL remains statistically evident over the past decade, though its strength and scope vary across game metrics. It manifests most strongly in scoring, yardage, and rushing attempts, and seems driven by cumulative factors like fan energy, routine, and familiarity. While the league continues to evolve—with neutral-site games, advanced travel planning, and more analytics-driven coaching—it’s clear that “playing at home” still matters.
